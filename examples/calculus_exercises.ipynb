## Exercise 4 - Find the global maximum and minimum for the following equation.

    f(x) = x^5 - 30(x)^3 + 50(x), where -5 <= x <= 5

```python
x = sympy.symbols('x')

f = x**5 - 30 * x**3 + 50 * x

d1 = sympy.diff(f, x)
print(f'First Derivative:\n\t{d1}\n')
d2 = sympy.diff(f, x, 2)
print(f'Second Derivative:\n\t{d2}\n')

# First Derivative Equal to Zero
d1_zeros = sympy.solve(d1, x)
print(f'First Derivative Equal to Zero:\n\t{d1_zeros}\n')

# Second Derivative Equal to Zero
d2_zeros = sympy.solve(d2, x)
print(f'Second Derivative Equal to Zero:\n\t{d2_zeros}\n')


set_min = float(d1_zeros[0])
print(f'\nSet Minimum:\n\tx = {set_min}\n')

set_max = float(d1_zeros[-1])
print(f'\nSet Maximum:\n\tx = {set_max}\n')


sympy.plot(f, (x, -5, 5), title=f'Function: {f}');
sympy.plot(d1, (x, -5, 5), title='First Derivative');
sympy.plot(d2, (x, -5, 5), title='Second Derivative');
```

## Exercise 8 - Create a funtion that uses gradient descent to find the minimum value of a single variable function. This also checks for the existence of a solution for the equation, f'(x)=0 and plots the intermediate points traversed. Explain the significance of differentiation.



```python
f = 3.5 * x**2 - 4.75 * x + 10
sympy.plot(f, (x, -4, 4));
```

```python
d1 = sympy.diff(f)
d1
```

```python
step = 0.1
tol = 0.01
iter_pts = []

prior = 0
guess = -4

while abs(guess - prior) > tol:
    prior = guess
    gradient = d1.subs(x, prior)
    guess = prior - (gradient * step)
    iter_pts.append(guess)

print(f'Local Minima: x = {iter_pts[-1]}')
```
