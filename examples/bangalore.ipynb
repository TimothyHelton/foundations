### Months Highest and Lowest Temperature

```python
def temp_min_max(year):
    """Find minimum and maximum months temperature given a year.
    
    :param int year: year to search
    :return: minimum and maximum monthly temperatures for a given year
    :rtype: str
    """
    y = data.loc[year, :]
    high_mon = y.idxmax()
    high_temp = y.max()
    low_mon = y.idxmin()
    low_temp = y.min()
    print(
        f'{year} Summary of High and Low Temperatures\n\n'
        f'High Month: {high_mon}, {high_temp:.2f}\n'
        f'Low Month: {low_mon}, {low_temp:.2f}'
    )
```

#### Month with Highest Temperatures for 1960

```python
temp_min_max(1960)
```

### Compare Distributions of Temperatures for Two Months

```python
aug = data.loc[:, 'Aug']
dec = data.loc[:, 'Dec']
combine = pd.concat([aug, dec])

axis_font = 14
title_font = 16
n_bins = np.linspace(math.floor(combine.min()), math.ceil(combine.max()), 20)
fig = plt.figure('Single Month Histogram', figsize=(12, 12), 
                 facecolor='white', edgecolor='black');
ax0 = plt.subplot2grid((2, 2), (0, 0));
ax1 = plt.subplot2grid((2, 2), (0, 1), sharey=ax0);
ax2 = plt.subplot2grid((2, 2), (1, 0), colspan=2, sharey=ax0);

ax0.hist(aug, n_bins, color='r', alpha=0.5, label='August');
ax0.set_title('August', fontsize=title_font)
ax0.set_xlabel('Temperature', fontsize=axis_font)
ax0.set_ylabel('Count', fontsize=axis_font)

ax1.hist(dec, n_bins, color='b', alpha=0.5, label='December');
ax1.set_title('December', fontsize=title_font)
ax1.set_xlabel('Temperature', fontsize=axis_font)

ax2.hist([aug, dec], n_bins, color=['r', 'b'], alpha=0.5, label=['August', 'December']);
ax2.set_xlabel('Temperature', fontsize=axis_font);
ax2.set_ylabel('Count', fontsize=axis_font);
plt.legend();

plt.suptitle('Histogram of Bangolore Temperatures\nAugust and December', fontsize=24);
```

### Compare Two Years in the Same Frame

### Function to Compare Two Months in the Same Frame

```python
data.loc[1979, 'Sep']
```

```python
def comp_months_hist(month_1, month_2):
    """Create overlaid histogram of two separate months.
    
    :param str month_1: 1st month abbreviation
    :param str month_2: 2nd month abbreviation 
    """
    mon_1 = data.loc[:, month_1]
    mon_2 = data.loc[:, month_2]
    combine = pd.concat([mon_1, mon_2])

    axis_font = 14
    title_font = 16
    n_bins = np.linspace(math.floor(combine.min()), math.ceil(combine.max()), 20)
    fig = plt.figure('Stacked Multi-Month Histogram', figsize=(8, 8), 
                     facecolor='white', edgecolor='black');
    ax0 = plt.subplot2grid((1, 1), (0, 0));

    ax0.hist([mon_1, mon_2], n_bins, histtype='barstacked', alpha=0.5,
             label=[month_1, month_2]);
    ax0.set_xlabel('Temperature', fontsize=axis_font)
    ax0.set_ylabel('Count', fontsize=axis_font)
    plt.legend()

    plt.suptitle(f'Histogram of Bangolore Temperatures\n{month_1} and {month_2}',
                 fontsize=24);
```

```python
comp_months_hist('Sep', 'Jul')
```

### Compare Two Months Using a Box Plot

```python
def comp_months_box(month_1, month_2):
    """Create box plot of two separate months.
    
    :param str month_1: 1st month abbreviation
    :param str month_2: 2nd month abbreviation 
    """
    mon_1 = data.loc[:, month_1].values
    mon_2 = data.loc[:, month_2].values

    axis_font = 14
    title_font = 16
    fig = plt.figure('Multi-Month Box Plot', figsize=(8, 8), 
                     facecolor='white', edgecolor='black');
    ax0 = plt.subplot2grid((1, 1), (0, 0));

    
    ax0.boxplot([mon_1, mon_2], labels=[month_1, month_2]);
    ax0.set_xlabel('Month', fontsize=14);
    ax0.set_ylabel('Temperature', fontsize=14);

    plt.suptitle('Box Plot of Bangolore Temperatures', fontsize=24);
```

```python
comp_months_box('Sep', 'Dec')
```

### Mak a KDE Plot of One Month

Each estimation has a distribution.

```python
sns.distplot(data.loc[:, 'Sep'], bins=20, hist=False, kde=True, rug=True);
```

### Compare Two Months Using KDE Plot

```python
def comp_months_kde(month_1, month_2=None, bins=20, histogram=False, rug=True):
    """Create a Kernel Density Estimate plot for one or two months.
    
    :param str month_1: first month of interest
    :param str month_2: second month of interest
    :param int bins: number of histogram bins
    :param bool histogram: if True a histogram will be included
    :param bool rug: if True a rug plot will be included
    """
    mon_1 = data.loc[:, month_1];
    n_bins = bins;
    if month_2:
        mon_2 = data.loc[:, month_2];
        combine = pd.concat([mon_1, mon_2]);
        n_bins = np.linspace(math.floor(combine.min()), math.ceil(combine.max()), bins);

    axis_font = 14;
    title_font = 16;
    fig = plt.figure('KDE Plot', figsize=(8, 8), 
                     facecolor='white', edgecolor='black');
    ax0 = plt.subplot2grid((1, 1), (0, 0));
    
    sns.distplot(data.loc[:, month_1], bins=n_bins, hist=histogram, kde=True, rug=rug, ax=ax0);
    if month_2:
        sns.distplot(data.loc[:, month_2], bins=n_bins, hist=histogram, kde=True, rug=rug, ax=ax0);
```

```python
comp_months_kde('Aug', 'Dec')
```

```python
comp_months_kde('Aug')
```

```python
comp_months_kde('Aug', 'Dec', histogram=True)
```

### Function to Compare Two Months on a Violin Plot

```python
def comp_months_violin(month_1, month_2):
    """Create a violin plot for two months.
    
    :param str month_1: first month of interest
    :param str month_2: second month of interest
    """
    mon_1 = data[month_1];
    mon_2 = data[month_2];

    axis_font = 14;
    title_font = 16;
    fig = plt.figure('Violin Plot', figsize=(8, 8), 
                     facecolor='white', edgecolor='black');
    ax0 = plt.subplot2grid((1, 1), (0, 0));
    
    sns.violinplot(data=[mon_1, mon_2], ax=ax0);
    plt.xticks([0, 1], [month_1, month_2])
    plt.suptitle('Violin Plot', fontsize=16)
```

```python
comp_months_violin('May', 'Oct')
```
